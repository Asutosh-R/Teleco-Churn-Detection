```python
churn_data= pd.read_csv('churn_data.csv')
churn_data.head()
```

```python
customer_data= pd.read_csv('customer_data.csv')
customer_data.head()
```

```python
internet_data= pd.read_csv('internet_data.csv')
internet_data.head()
```

```python
merged= pd.merge(churn_data, customer_data,how='inner',on = 'customerID')
merged.head()
```

```python
teleco= pd.merge(merged,internet_data,how= 'inner', on= 'customerID')
teleco.head()
```

```python
churn= teleco['Churn']
teleco.drop('Churn',axis=1,inplace= True)
```

```python
teleco['Churn']= churn
```

```python
teleco.columns
```

```python
teleco.nunique()
```

```python
varlist= ['PhoneService','PaperlessBilling','Partner','Dependents','Churn']

def binary_con(x):
    return x.map({'Yes': 1,'No':0})

teleco[varlist] = teleco[varlist].apply(binary_con)
    
```
